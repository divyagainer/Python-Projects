```python
def sample_function(row):
    cost=0


    counter = row['Weight_slab_metric']
    # if counter == 1 :
    if counter == 1 & bool(row['Type of Shipment']=='Forward charges'):

        cost = row['fwd_fixed']
        status = bool(row['Type of Shipment']=='Forward charges')
        # print()
        return cost


    elif counter == 1 &  bool(row['Type of Shipment']=='Forward and RTO charges'):

        cost = row['rto_fixed']
        return cost

    elif counter > 1 & bool(row['Type of Shipment']=='Forward charges'):

        cost_a=0
        cost_b=0
        final_cost=0
        cost_a = row['fwd_fixed']
        cost_b = row['fwd_additional'] * (counter-1)
        final_cost= cost_a + cost_b
        return final_cost

    else:

        cost_a=0
        cost_b=0
        final_cost=0
        cost_a = row['rto_fixed']
        cost_b = row['rto_additional'] * (counter-1)
        final_cost = cost_a + cost_b
        return final_cost
```
